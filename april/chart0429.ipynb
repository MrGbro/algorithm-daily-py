```python
from typing import List


class Solution2962:
    def countSubarrays(self, nums: List[int], k: int) -> int:
        max_v = 0
        for v in nums:
            max_v = max(max_v, v)
        cnt = 0
        l = 0
        res = 0
        for v in nums:
            if v == max_v:
                cnt += 1
            while cnt >= k:
                d = nums[l]
                if d == max_v:
                    cnt -= 1
                l += 1
            res += l
        return res


inst = Solution2962()
count = inst.countSubarrays([1, 3, 2, 3, 3], 2)
print(count)
```

```python
from typing import Optional


class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class Solution328:
    def oddEvenList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if not head or not head.next:
            return head
        dummy = ListNode(-1)
        cur = dummy
        slow = head
        fast = head.next
        while slow:
            cur.next = ListNode(slow.val)
            cur = cur.next
            if not slow.next:
                break
            slow = slow.next.next
        while fast and fast.next:
            cur.next = ListNode(fast.val)
            cur = cur.next
            if not fast.next:
                break
            fast = fast.next.next
        return dummy.next
```

```python
class Solution86:
    def partition(self, head: Optional[ListNode], x: int) -> Optional[ListNode]:
        pre = ListNode(-1)
        below = ListNode(-1)
        pre_cur = pre
        below_cur = below
        while head:
            tm = head
            head = head.next
            tm.next = None
            if tm.val < x:
                pre_cur.next = tm
                pre_cur = pre_cur.next
            else:
                below_cur.next = tm
                below_cur = below_cur.next
        pre_cur.next = below.next
        return pre.next
```

```python
class Solution160:
    def getIntersectionNode(
        self, headA: ListNode, headB: ListNode
    ) -> Optional[ListNode]:
        listA = headA
        listB = headB
        while listA != listB:
            listA = headB if not listA else listA.next
            listB = headA if not listB else listB.next

        return listA
```

```python
class Solution2:
    def addTwoNumbers(
        self, l1: Optional[ListNode], l2: Optional[ListNode]
    ) -> Optional[ListNode]:
        dummy = ListNode(-1)
        cur = dummy
        carry = 0
        while l1 and l2:
            s = carry + l1.val + l2.val
            cur.next = ListNode(s % 10)
            carry = s // 10
            l1 = l1.next
            l2 = l2.next
            cur = cur.next
        while l1:
            s = carry + l1.val
            cur.next = ListNode(s % 10)
            carry = s // 10
            l1 = l1.next
            cur = cur.next
        while l2:
            s = carry + l2.val
            cur.next = ListNode(s % 10)
            carry = s // 10
            l2 = l2.next
            cur = cur.next
        if carry != 0:
            cur.next = ListNode(carry)
        return dummy.next
```

```python
class Solution:
    def addTwoNumbers(
        self, l1: Optional[ListNode], l2: Optional[ListNode]
    ) -> Optional[ListNode]:
        def rev(node: Optional[ListNode]) -> Optional[ListNode]:
            if not node or not node.next:
                return node
            reverse = rev(node.next)
            node.next.next = node
            node.next = None
            return reverse

        dummy = None
        nl1 = rev(l1)
        nl2 = rev(l2)

        carry = 0
        while nl1 and nl2:
            s = nl1.val + nl2.val + carry
            carry = s // 10
            n = ListNode(s % 10)
            n.next = dummy
            dummy = n
            nl1 = nl1.next
            nl2 = nl2.next

        while nl1:
            s = nl1.val + carry
            carry = s // 10
            n = ListNode(s % 10)
            n.next = dummy
            dummy = n
            nl1 = nl1.next

        while nl2:
            s = nl2.val + carry
            carry = s // 10
            n = ListNode(s % 10)
            n.next = dummy
            dummy = n
            nl2 = nl2.next
        if carry != 0:
            n = ListNode(carry)
            n.next = dummy
            dummy = n
        return dummy
```

```python
class Solution2816:
    def doubleIt(self, head: Optional[ListNode]) -> Optional[ListNode]:
        def rev(node: Optional[ListNode]) -> Optional[ListNode]:
            if not node or not node.next:
                return node
            reverse = rev(node.next)
            node.next.next = node
            node.next = None
            return reverse

        nh = rev(head)
        dummy = None
        carry = 0
        while nh:
            s = nh.val * 2 + carry
            n = ListNode(s % 10)
            carry = s // 10
            n.next = dummy
            dummy = n
            nh = nh.next
        if carry != 0:
            n = ListNode(carry)
            n.next = dummy
            dummy = n
        return dummy
```

```python
class Solution21:
    def mergeTwoLists(
        self, list1: Optional[ListNode], list2: Optional[ListNode]
    ) -> Optional[ListNode]:
        dummy = ListNode(-1)
        cur = dummy
        while list1 and list2:
            if list1.val < list2.val:
                cur.next = list1
            else:
                cur.next = list2
            cur = cur.next
        while list1:
            cur.next = list1
            cur = cur.next
        while list2:
            cur.next = list2
            cur = cur.next
        return dummy.next
```

```python
import heapq


class Solution23:
    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        priority_queue = []
        for v in lists:
            if v:
                heapq.heappush(priority_queue, (v.val, v))
        dummy = ListNode(-1)
        cur = dummy
        while len(priority_queue) > 0:
            prio, item = heapq.heappop(priority_queue)
            cur.next = item
            cur = cur.next
            nxt = item.next
            if nxt:
                heapq.heappush(priority_queue, (nxt.val, nxt))

        return dummy.next
```

```python
class Solution148:
    def sortList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        dummy = ListNode(-1)
        cur = dummy
        prio = []
        while head:
            heapq.heappush(prio, head.val)
            head = head.next
        while len(prio) > 0:
            cur.next = ListNode(heapq.heappop(prio))
            cur = cur.next
        return dummy.next
```

```python
from collections import deque


class Solution1019:
    def nextLargerNodes(self, head: Optional[ListNode]) -> List[int]:
        def reverse(node: Optional[ListNode]) -> Optional[ListNode]:
            if not node or not node.next:
                return node
            rever = reverse(node.next)
            node.next.next = node
            node.next = None
            return rever

        rev = reverse(head)
        res = deque()
        stack = deque()
        while rev:
            while len(stack) > 0 and stack[0] <= rev.val:
                stack.popleft()
            if len(stack) == 0:
                res.appendleft(0)
            else:
                res.appendleft(stack[0])
            stack.appendleft(rev.val)
            rev = rev.next
        return list(res)


inst = Solution1019()
h = ListNode(2, ListNode(1, ListNode(5)))
result = inst.nextLargerNodes(h)
print(result)
```
