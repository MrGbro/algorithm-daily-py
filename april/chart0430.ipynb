```python
from typing import Optional


class ListNode:
    def __init__(self, val=0, next: "ListNode" = None):
        self.val = val
        self.next = next


class Solution1171:
    def removeZeroSumSublists(self, head: Optional[ListNode]) -> Optional[ListNode]:
        pre_sum = 0
        map_dic = {}
        dummy = ListNode(0, head)
        cur = dummy
        while cur:
            pre_sum += cur.val
            map_dic[pre_sum] = cur
            cur = cur.next

        pre_sum = 0
        cur = dummy
        while cur:
            pre_sum += cur.val
            cur.next = map_dic[pre_sum].next
        return dummy.next
```

```python
from typing import List


class Solution:
    def findNumbers(self, nums: List[int]) -> int:
        result: int = 0
        for v in nums:
            cnt = 0
            while v > 0:
                cnt += 1
                v = v // 10
            result += 1 if cnt % 2 == 0 else 0

        return result
```

```python
class ListNode:
    def __init__(self, val: int, next=None):
        self.val = val
        self.next = next


class MyLinkedList:

    def __init__(self):
        self.size = 0
        self.head = ListNode(0)

    def get(self, index: int) -> int:
        if index >= self.size or index < 0:
            return -1
        cur = self.head
        for i in range(index + 1):
            cur = cur.next
        return cur.next

    def addAtHead(self, val: int) -> None:
        self.addAtIndex(0,val)

    def addAtTail(self, val: int) -> None:
        self.addAtIndex(self.size,val)

    def addAtIndex(self, index: int, val: int) -> None:
        if index > self.size:
            return
        index = max(index,0)
        cur = self.head
        for i in range(index):
            cur = cur.next
        to_add = ListNode(val)
        self.size += 1
        to_add.next = cur.next
        cur.next = to_add

    def deleteAtIndex(self, index: int) -> None:
        if index <0 or index >= self.size:
            return
        cur = self.head
        for i in range(index):
            cur = cur.next
        self.size -= 1
        cur.next = cur.next.next


# Your MyLinkedList object will be instantiated and called as such:
obj = MyLinkedList()
obj.addAtHead(1)
obj.addAtTail(3)
obj.addAtIndex(1, 2)
param_1 = obj.get(1)
obj.deleteAtIndex(1)
delete_v = obj.deleteAtIndex(1)
```
