```python
class ListNode:
    def __init__(self, val: int, next: 'ListNode' = None):
        self.val = val
        self.next = next


class MyLinkedList:

    def __init__(self):
        self.size = 0
        self.head = ListNode(0)

    def get(self, index: int) -> int:
        if index >= self.size or index < 0:
            return -1
        cur = self.head
        for i in range(index + 1):
            cur = cur.next
        return cur.val

    def addAtHead(self, val: int) -> None:
        self.addAtIndex(0, val)

    def addAtTail(self, val: int) -> None:
        self.addAtIndex(self.size, val)

    def addAtIndex(self, index: int, val: int) -> None:
        if index > self.size:
            return
        idx = max(index, 0)
        cur = self.head
        for i in range(idx):
            cur = cur.next
        add_node = ListNode(val)
        add_node.next = cur.next
        cur.next = add_node
        self.size += 1

    def deleteAtIndex(self, index: int) -> None:
        if index >= self.size or index < 0:
            return
        cur = self.head
        for i in range(index):
            cur = cur.next
        cur.next = cur.next.next
        self.size -= 1
```

```python
#LeetCode146
class LRUCache:

    def __init__(self, capacity: int):
        self.cap = capacity
        self.container = {}

    def get(self, key: int) -> int:
        if key in self.container:
            val = self.container[key]
            del self.container[key]
            self.container[key] = val
            return val
        return -1

    def put(self, key: int, value: int) -> None:
        if key in self.container:
            del self.container[key]
            self.container[key] = value
        else:
            while len(self.container) >= self.cap:
                first_key = next(iter(self.container.keys()))
                del self.container[first_key]
            self.container[key] = value
```
